#Inverse Laplace transform function 
The functions defined here are used to symbolically evaluate the inverse Laplace transform with the method described in Appendix A

```python
def ivt(uhat,s,n):
  ''' 
  ------------------------------------------------------------------------------
  Extension of the initial value theorem which computes the n'th derivative of 
  u(t) evaluated at t=0 using the Laplace transform of u(t), uhat(s).
  
  This is equation A.5 in Appendix A
                                                                                              
  PARAMETERS                                                                                                       
  ----------                                                                                                       
    uhat: Laplace transform of u. This is a symbolic expression containing s                                                
    s: Laplace domain variable                                                                                 
    n: the derivative order                                                                                        
                                                                                                                   
  RETURNS                                                                                                          
  -------                                                                                                          
    u_n:  symbolic expression for the nth derivative of u evaluated at t=0                                                                                        
  '''
  if n == 0:
    expr = s*uhat
    u_0 = expr.limit(s,np.inf)
    return u_0

  elif n > 0:      
    expr = s**(n+1)*uhat - sum(s**m*ivt(uhat,s,n-m) for m in range(1,n+1))
    u_n = expr.limit(s,np.inf)  
    return u_n

def ilt(uhat,s,t,N):
  '''   
  ------------------------------------------------------------------------------
  Evaluates the inverse Laplace transform of uhat(s) through a Taylor series 
  expansion
   
  This is a combination of Equation A.7 and A.5 in Appendix A                                                                                                   
                                                                                                                   
  PARAMETERS                                                                                                       
  ----------                                                                                                       
    uhat: Laplace transform of u. This is a symbolic expression containing s                                                    
    s: Laplace domain variable
    t: time domain variable
    N: order of the Taylor series expansion                                                                        
                                                                                                                   
  RETURNS                                                                                                          
  -------                                                                                                          
    series: symbolic expression for the series expansion of the u(x)                                               
          about x=0                                                                                                
  '''
  series = sum((ivt(uhat,s,n)*t**n)/sp.factorial(n) for n in range(N+1))  
  return series
```

## Two-dimensional Two layererd earthquake model
I consider here a two-dimensional two layered earthquake model.  I demonstrate that the initial rate of surface deformation resulting from viscoelastic relaxation is linear with respect to the fluidities of the two layers, regardless of whether the shear modulus is homogeneous in the halfspace.

##Variables used                                                                                                     
  x: distance from fault                                                                                     
  t: time                                                                                                          
  D: locking depth of the fault                                                                                    
  H: thickness of the top layer                                                                                    
  mu1,mu2: shear modulus of the top layer and lower substrate                                                             
  phi1,phi2: fluidity (1/viscosity) of the top layer and lower substrate 

```python
# define shear modulii parameters
mu1,mu2 = sp.symbols('mu1,mu2')
# define viscosity parameters
eta1,eta2 = sp.symbols('eta1,eta2')
# define fluidity parameters 
phi1,phi2 = sp.symbols('phi1,phi2')
etainv1,etainv2,etainv3 = sp.symbols('eta1^-1,eta2^-1,eta3^-1')

# time, dummy variable for time integration, and Laplace domain parameter
t,theta,s = sp.symbols('t,theta,s')

# slip function.  This needs to be explicitly defined and I define it 
# here as one unit of slip throughout the postseismic period
b = 1

# dummy summation variable
n = sp.symbols('n')

# distance from the fault
x = sp.symbols('x')

# locking depth and upper layer thickness
D,H = sp.symbols('D H')

# number of terms to approximate the infinite series
nmax = 3

# W is equation 3 in the text and it is defined here as a function 
# rather than an explicit expression for the sake of cleaner output 
W = sp.Function('W')(n)

# equation 4 in text
gamma = (mu1-mu2)/(mu1+mu2)

# equation 2 in text. This describes the elastic surface displacements
u_e = b*(sp.Rational(1,2)*W.subs(n,0) + sp.summation(gamma**n*W,(n,1,nmax)))

# equation 5 in text.  This is the Laplace transform of equation 2
uhat_e = sp.laplace_transform(u_e,t,s)[0]

# equation 8 in text. These are the shear modulii in the Laplace domain 
# for a Maxwell viscoelastic material. Note I am using fluidity instead of 
# viscosity here
mu1hat = s/((s/mu1) + phi1)
mu2hat = s/((s/mu2) + phi2) 

# equation 6 in text. This is the Laplace transform of the viscoelastic
# surface displacements truncated after two terms (i.e. containing initial 
# displacement and initial rate of displacement)
uhat_v = uhat_e.subs(((mu1,mu1hat),(mu2,mu2hat)))

# This is the surface displacement in the time domain for the viscoelastic 
# lithosphere without assuming that mu1 == mu2.  Although the solution is long,
# one can recognize the initial displacements resulting from the elastic
# dislocation, followed by the initial rate of displacement which is indeed 
# linear with respect to the fluidity in each layer, phi1 and phi2.  
u_v = ilt(uhat_v,s,t,1)
u_v.expand().collect((phi1,phi2))
```

```python
# If we assuming that the shear modulii are equal then we will recover 
# eq. 10 in the text, with the exception that b(t) is explicitly defined here
u_v = ilt(uhat_v.subs(mu2,mu1),s,t,2)
u_v
```

##Two-dimensional three layered earthquake model
I considered here a three layered two-dimensional viscoelastic earthquake model.  I will demonstrate how we arrrived at equation 14 in the text.how we arrived at eq 14 in the text.  We use the solution for the three layered elastic solution from Chinnery and Jovanovich 1972. We adopt the numbering scheme used in Chinnery and Jovanovich and described below.  Note that equations 12 through 14 in the text use a different numbering scheme.                                
                                                                                                                   
##variables used: 
  x: distance from fault                                                                                     
  t: time 
  mu3,mu2,mu1: shear modulus for the top, middle, and bottom layer (mu4 is assumed 0)                                                               
  phi3,phi2,phi1: fluidity (1/viscosity) for the top, middle, and bottom layer                                                  
  h3,h2: thichness of the top and middle layer                                                                     
  p: depth to the bottom of the fault. (down is positive)
  

```python
# define shear modulii parameters
mu1,mu2,mu3 = sp.symbols('mu1,mu2,mu3')

# define fluidity parameters
phi1,phi2,phi3 = sp.symbols('phi1,phi2,phi3')

# time, dummy variable for time integration, and Laplace domain parameter
t,theta,s = sp.symbols('t,theta,s')

# slip function.  This needs to be explicitly defined and I define it 
# here as one unit of slip throughout the postseismic period
b = 1

# dummy summation variable
M,N = sp.symbols('M,N')

# distance from the fault
x = sp.symbols('x')

# locking depth
p = sp.symbols('p')

# equation 13 in text which is only expressed as a function for cleaner output
W = sp.Function('W')(M,N)

# Chinnery and Jovanovich 1972 solution
l_max = 2
m_max = 2
n_max = 2

a1    = (mu2 - mu1) / (mu2 + mu1)
a2    = (mu3 - mu2) / (mu3 + mu2)
a3    = -1
b1    = 2*mu1 / (mu1 + mu2)
b2    = 2*mu2 / (mu2 + mu3)
b3    = 2
d1    = 2*mu2 / (mu1 + mu2)
d2    = 2*mu3 / (mu2 + mu3)

def P(l,m,n):
  N = sp.factorial(n+1) * sp.factorial(l+m-1)
  D = sp.factorial(l) * sp.factorial(n) * sp.factorial(l-1) * sp.factorial(m)
  return N / D

def Q(l,m,n):
  N = sp.factorial(n+l) * sp.factorial(l+m)
  D = sp.factorial(l) * sp.factorial(n) * sp.factorial(l) * sp.factorial(m)
  return N / D

u = -b3*sp.Rational(-1,2)*W.subs(((N,0),(M,0)))
for n in range(n_max):
  coeff = a2*b3*(-a2*a3)**n
  u += coeff*W.subs(((M,0),(N,n+1)))

for l in range(1,l_max):
  for m in range(m_max):
    for n in range(n_max):
      coeff = a2*b3*(-a1*a2)**m*(-a2*a3)**n*(-a2*a3*d2*b2)**l*P(l,m,n)
      u += coeff*W.subs(((M,l+m),(N,l+n+1)))

for l in range(l_max):
  for m in range(m_max):
    for n in range(n_max):
      coeff = a1*d2*b2*b3*(-a1*a3)**m*(-a2*a3)**n*(-a1*a3*d2*b2)**l*Q(l,m,n)
      u += coeff*W.subs(((M,l+m+1),(N,l+n+1)))
     
u_e = sp.Rational(1,2)*b*u

# This is the laplace transform of the elastic displacements
uhat_e = sp.laplace_transform(u_e,t,s)[0]

# define the equivalent shear modulii for a Maxwell viscoelastic body
mu1hat = s/((s/mu1) + (phi1))
mu2hat = s/((s/mu1) + (phi2))
mu3hat = s/((s/mu1) + (phi3))

# substitute the equivalent shear modulii into the Laplace transform of the 
# elastic displacements to get the laplace transform of the viscoelatic 
# displacements
uhat_v = uhat_e.subs(((mu1,mu1hat),(mu2,mu2hat),(mu3,mu3hat)))

# take the inverse Laplace transform to find the viscoelastic displacements
# in the time domain. I am only expanding to the first two terms.  We 
# recover equation 14 in the text if you note that slip is explicitly defined,
# and the numbering of the layers is reversed here.
u_v = ilt(uhat_v,s,t,1)
u_v
```
